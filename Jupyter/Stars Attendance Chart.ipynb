```python
from trout.stars import get_star, STAR_START, STAR_END

import matplotlib.pyplot as plt
```

```python
DEFAULT_YEARS = [
    2003, 2005, 2006, 2007, 2008, 
    2009, 2010, 2011, 2012, 2013, 
    2014, 2015, 2016, 2017, 2019, 
    2020, 2021]
```

```python
def stars_attendance( year=None, 
                    start=STAR_START, 
                    end=STAR_END,
                    ignore_star=lambda x : False,
                    show_chart=True):
    """Calculates attendance of stars in given range for a given year or all years"""
    
    # Dict to store star attendances
    attendances = {}
    for star_no in range(start, end + 1):
        # ignore_star function is useful when we try to cherry-pick 
        # which stars to find attendance of 
        if ignore_star(star_no):
            continue
        # Note get_star(star_no).attendance(None) means to calc att. for entire dataset
        attendances[star_no] = get_star(star_no).attendance(year) 
    
    if show_chart:
        x, y = zip(*attendances.items())
        plt.plot(x, y, 'b+')
        plt.xlabel('Star #')
        plt.ylabel('% Attendance')
        if not year:
            plt.title(f"Overall Attendance for Stars {start}-{end}")
        else:
            plt.title(f"{year} Attendance for Stars {start}-{end}")
        plt.show()
    
    return attendances
```

```python
def star_attendance(star_no, years=DEFAULT_YEARS, show_chart=True):
    """Shows char for attendance of a star over our years range"""
    attendances = {}
    star = get_star(star_no)
    for year in years:
        attendances[year] = star.attendance(year)
    
    if show_chart:
        x, y = zip(*attendances.items())
        plt.plot(x, y, 'b+')
        plt.xlabel('Year')
        plt.ylabel('% Attendance')
        plt.title(f"Star {star_no} Attendance")
        plt.show()    
        
    return attendances
```
